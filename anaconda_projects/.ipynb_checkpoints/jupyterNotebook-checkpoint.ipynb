### OpenCV 의 영상은 칼라순서가 RGB 가 아니라 BGR로 되어 있음

````python
img = cv2.imread("test.jpg") 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR을 RGB로 다시 바꾸어줌
plt.imshow(img)
plt.show()
````
